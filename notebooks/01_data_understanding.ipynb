## Target Variable Notes

The project will create a binary default target from `loan_status`.

Planned target mapping:

- `Fully Paid` = 0
- `Charged Off` = 1
- `Default` = 1
- `Late (31-120 days)` = 1

Loans with statuses such as `Current`, `In Grace Period`, or incomplete outcomes will be excluded from supervised model training because the final repayment outcome is not yet known.

```python
def map_default_status(status):
    if status in ["Charged Off", "Default", "Late (31-120 days)"]:
        return 1
    elif status == "Fully Paid":
        return 0
    else:
        return np.nan

df["default_flag"] = df["loan_status"].apply(map_default_status)

df["default_flag"].value_counts(dropna=False)
```

```python
df_model = df[df["default_flag"].notna()].copy()
df_model["default_flag"] = df_model["default_flag"].astype(int)

df_model.shape
```

## Leakage Review

Potential leakage variables are fields that would only be known after loan origination or during repayment. These variables may reveal whether the borrower repaid, defaulted, entered hardship, settled the loan, or made payments.

These fields will be excluded from model training to keep the project realistic for loan application risk scoring.

```python
candidate_features = [
    "loan_amnt",
    "funded_amnt",
    "term",
    "int_rate",
    "installment",
    "grade",
    "sub_grade",
    "emp_length",
    "home_ownership",
    "annual_inc",
    "verification_status",
    "purpose",
    "dti",
    "delinq_2yrs",
    "earliest_cr_line",
    "inq_last_6mths",
    "open_acc",
    "pub_rec",
    "revol_bal",
    "revol_util",
    "total_acc",
    "application_type",
    "mort_acc",
    "pub_rec_bankruptcies",
    "issue_d"
]

existing_candidate_features = [col for col in candidate_features if col in df_model.columns]

existing_candidate_features
```

```python
candidate_feature_df = pd.DataFrame({
    "column": existing_candidate_features,
    "reason_for_use": [
        "Candidate origination-time feature" for _ in existing_candidate_features
    ]
})

candidate_feature_df.to_csv("../data/outputs/candidate_model_features.csv", index=False)

candidate_feature_df
```

## Key Findings
